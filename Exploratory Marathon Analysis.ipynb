**What is the difference in speed for 50km, and 50mi races male to female?**

```python
df2.groupby(['race_length', 'athlete_gender'])['athlete_avg_speed'].mean()
```

***Male vs Female:*** *For 50km races, around 0.7 difference, for 50mi races it is around a 0.5 difference*

**What age groups perform the best in 50 Mile races?
(minimum of 20+ races)**

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
```

*Racers in their 20's perform the best with a minimum of 20+ races in their career*

```python
df2.query('race_length == "50mi"').groupby('athlete_age')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>19')
```

**Do racers perform better in the summer or the winter?**

* *Spring: Mar-May 3-5*
* *Summer: Jun-Aug 6-8*
* *Fall: Sep-Nov 9-11*
* *Winter: Dec-Feb 12-2*

```python
df2['race_month'] = df2['race_date'].str.split('.').str.get(1).astype(int)
```

```python
df2['race_season'] = df2['race_month'].apply(lambda x: 'Winter' if x > 11
                                            else 'Fall' if x > 8
                                            else 'Summer' if x > 5
                                            else 'Spring ' if x > 2
                                            else 'Winter')
```

```python
df2.groupby('race_season')['athlete_avg_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

*It's easier to run a race in the winter than it is to run a race in the summer.*
