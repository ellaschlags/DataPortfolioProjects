### Male Births vs Female Births

```python
plt.figure(figsize=(12,6))

plt.plot(df['year'],df['birth_female'],label='Birth female')
plt.plot(df['year'],df['birth_male'], label='Birth male')

plt.legend()
plt.xlabel('Year')
plt.ylabel('No. of births')
plt.title('Male child vs Female child births')

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df['year'],df['stillbirth_female'],label='Female stillbirths')
plt.plot(df['year'],df['stillbirth_male'], label='Male stillbirths')

plt.legend()
plt.xlabel('Year')
plt.ylabel('No. of deaths')
plt.title('Male vs Female stillbirths')

plt.show()
```

```python
malePercentage = round(df['population_male'].mean()/df['population_total'].mean()*100,2)
femalePercentage = round(df['population_female'].mean()/df['population_total'].mean()*100,2)

plt.pie([malePercentage,femalePercentage],labels=['Males','Females'],autopct = '%1.1f%%',
        startangle = 90,
        wedgeprops = {"edgecolor" : "black",
                      'linewidth': 2,
                      'antialiased': True})
plt.axis('equal')
plt.title('Mean male-female ratio')
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2)

ax1.barh(df["year"], df["stillbirth_male"], align='center')
ax2.barh(df["year"], df["stillbirth_female"], align='center', color="orange")

ax1.invert_xaxis()
ax1.set_title("Male Stillbirths")
ax2.set_title("Female Stillbirths")

ax1.yaxis.set_ticklabels([])

plt.show()
```

### Mother's age vs Father's age

```python
plt.figure(figsize=(12,6))
bubbleSize = 20

plt.scatter(df['year'],df['mother_age_firstborn'],bubbleSize)
plt.plot(df['year'],df['mother_age_firstborn'],label="Mother's age when firstborn is born")

plt.scatter(df['year'],df['mother_age_secondborn'],bubbleSize)
plt.plot(df['year'],df['mother_age_secondborn'], label="Mother's age when secondborn is born")

plt.scatter(df['year'],df['mother_age_thirdborn'],bubbleSize)
plt.plot(df['year'],df['mother_age_thirdborn'], label="Mother's age when thirdborn is born")

plt.legend()
plt.xlabel('Year')
plt.ylabel('Age')
plt.title("Mother's age when having kids")
plt.show()
```

```python
plt.figure(figsize=(12,6))
bubbleSize = 20

plt.scatter(df['year'],df['father_age_firstborn'],bubbleSize)
plt.plot(df['year'],df['father_age_firstborn'],label="Father's age when firstborn is born")

plt.scatter(df['year'],df['father_age_secondborn'],bubbleSize)
plt.plot(df['year'],df['father_age_secondborn'], label="Father's age when secondborn is born")

plt.scatter(df['year'],df['father_age_thirdborn'],bubbleSize)
plt.plot(df['year'],df['father_age_thirdborn'], label="Father's age when thirdborn is born")

plt.legend()
plt.xlabel('Year')
plt.ylabel('Age')
plt.title("Father's age when having kids")
plt.show()
```
